**IMPORT LIBRARY MODEL**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
df['Efficiency'] = df['ROI'].apply(lambda x: 'Efisien' if x >= 1.5 else 'Tidak Efisien')

features = ["Budget", "Clicks", "Conversions", "Revenue", "Discount", "Units_Sold"]
target = 'Efficiency'

X = df[features]
y = df[target]

NIM = "1305223027"
random_state = int(NIM[-3:])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
```

**DATA UJI & EVALUASI MODEL**

```python
model = RandomForestClassifier(random_state=random_state)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label='Efisien')
recall = recall_score(y_test, y_pred, pos_label='Efisien')
f1 = f1_score(y_test, y_pred, pos_label='Efisien')
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
```

**FEATURE IMPORTANCE**

```python
if isinstance(model, RandomForestClassifier):
    feature_importance = pd.Series(model.feature_importances_, index=features).sort_values(ascending=False)
    print("\nFeature Importance:")
    print(feature_importance)
```

Pada hasil feature importance tersebut, **budget** merupakan fitur dengan kepentingan tertinggi dan memiliki pengaruh tinggi pada kampanye produk

**VISUALISASI CONFUSION MATRIX**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Tidak Efisien', 'Efisien'], yticklabels=['Tidak Efisien', 'Efisien'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
